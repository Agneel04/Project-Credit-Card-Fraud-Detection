CREDIT CARD FRAUD DETECTION USING DATASET FROM KAGGLE

IMPORTING LIBRARIES REQUIRED

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split #MACHINE LEARNING MODEL FOR REGRESSION AND OTHER TOOLS
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_dataset=pd.read_csv('/content/creditcard.csv')
```

```python
credit_card_dataset.head()
```

```python
credit_card_dataset.tail()
```

```python
credit_card_dataset.info() #info about the dataset
```

```python
credit_card_dataset.isnull().sum()
```

No Null Available

```python
credit_card_dataset['Class'].value_counts()
```

```python
legit = credit_card_dataset[credit_card_dataset.Class == 0]
fraud = credit_card_dataset[credit_card_dataset.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

Class group by mean

```python
credit_card_dataset.groupby('Class').mean()
```

Taking a sample of 249 since the dataset is uneven

```python
legit_sample = legit.sample(n=249)
```

New dataset with 249 legit and 249 fraud datas

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Spliting into Features and targets Class-Target....

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
print(X,Y)
```

Spliting the data into taraining and testing data with 0.8->80% train size and 0.3->20% into test size and randomly catagory 2 with Stratifying Y    **%->percentile**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.22, stratify=Y, random_state=4)
print(X.shape, X_train.shape, X_test.shape)
```

Model Training and Logistic Regression

```python
# training the Logistic Regression Model with Training Data
model = LogisticRegression()
model.fit(X_train, Y_train)
```

Accuracy of Training Set

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print(training_data_accuracy)
```

Accuracy of Testing Set

```python
X_test_prediction = model.predict(X_test)
training_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print(training_data_accuracy)
```
